```python
## Import the Pandas and Seaborne Libraries

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sd = pd.read_csv("D:/Python Projects/Data visualizations using Seaborne/Shooting_Data.csv")
```

```python
sd.head()
```

```python
sns.countplot(x='gender',data=sd)
plt.show()

## Result - The number of males involved in Police shootings is significantly higher as compared to females
```

```python
sns.countplot(x='race',data=sd)
plt.show()

## Result - White people were mostly involved in Police shootings followed by Black and Hispanic people
```

```python
sns.countplot(x='gender',data=sd,hue='signs_of_mental_illness')
plt.show()

## For a given gender, the number of people with NO mental illness was higher as compared to people with mental illness
```

```python
df = sd[sd['state'] == 'CA']
sns.countplot(x='race',data=df,hue='gender')
plt.title('Shooting stats in the state of California')
plt.show()

## In the state of California, the number of hispanic males were mostly involved in the police shooting 

## In the state of California, the number of white females were the mostly involved in the police shooting 
```

```python
## In the police shooting stats, plot the top 10 states with cases of of mental illness?

mi = sd[sd['signs_of_mental_illness'] == True]

mi.head()

df_mi = mi.groupby('state')[['signs_of_mental_illness']].count()

df_mi.sort_values(by=['signs_of_mental_illness'],ascending=False).head(10)

## The state of California had highes number of people identified with mental illness followed by state of Florida
```

```python
df_sd = sd.groupby('age')[['name']].count().sort_values('name',ascending=False)

df_sd.head(20)


## Most number of people in the police shootout were aged 31.
```

```python
sd.head()
```

```python
## Add a new column called year

sd['year'] = pd.DatetimeIndex(sd['date']).year
sd.head()
```

```python
sd.groupby('year')[['name']].count().head(10)

```

```python
sns.countplot(x = 'year', data = sd)
plt.title('Police shootout stats from year 2015 to 2020')
plt.show()
```

```python
sns.relplot(x='year',y='age',data=sd,kind='line')
plt.show()

## The mean age of a person in the police shootout has increased from 2015 to 2019. 
```

```python
sns.set_style("whitegrid")

sns.set_palette("RdBu")

sns.set_context("talk")

x=sns.catplot(x='threat_level',data=sd,kind="count")

plt.xticks(rotation=0)

x.fig.suptitle('Comparison of threat level')

x.set(xlabel="Threat Level")
plt.show()


## People identified in police shootout were considered to be attacking in nature
```
